##  Importing and Saving the Movies Dataset (Best Practice)

```python
import pandas as pd
import requests
import json
pd.options.display.max_columns = 30
```

```python
movie_id = [0, 299534, 19995, 140607, 299536, 597, 135397,
            420818, 24428, 168259, 99861, 284054, 12445,
            181808, 330457, 351286, 109445, 321612, 260513]
```

```python
basic_url = 'https://api.themoviedb.org/3/movie/{}?{}' 
```

```python
json_list = []
for movie in movie_id:
    url = basic_url.format(movie, api_key)#replace placeholder in basic-url
    r = requests.get(url)#response object
    if r.status_code != 200: # data is available if status code is 200
        continue#status code not equal to 200 - stop the iteration and go to the next movie id
    else:
        data = r.json()
        json_list.append(data) 
df = pd.DataFrame(json_list)
```

```python
requests.get(basic_url.format(0, api_key)).status_code #404 no data available
```

```python
df
```

```python
#overwrite df, sort revenue by high to low
df = df.loc[:, ["title", "id", "revenue", "genres", "belongs_to_collection", "runtime"]].sort_values(by = "revenue",
                                                                                                ascending = False)
```

```python
df
```

```python
#save and store our dataset
#export it to a json file
#best to handle nested data
df.to_json("movies.json", orient = "records")
```
